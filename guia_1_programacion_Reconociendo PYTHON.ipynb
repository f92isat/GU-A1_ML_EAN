## Determinación de los tipos de variables

En esta sección, identificaremos el tipo de cada variable en el DataFrame para entender la naturaleza de los datos.


```python
import pandas as pd

# Cargar la base de datos Titanic desde un repositorio en línea
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df = pd.read_csv(url)

# Función para clasificar las variables
def clasificar_tipo_variable(df):
    tipo_variable = {}

    for columna in df.columns:
        unique_values = df[columna].nunique()
        if df[columna].dtype == 'object':
            if unique_values < 10:
                # Variables categóricas (cualitativas)
                if df[columna].nunique() == df[columna].rank(method='first').nunique():
                    tipo_variable[columna] = 'Cualitativa Ordinal'
                else:
                    tipo_variable[columna] = 'Cualitativa Nominal'
            else:
                tipo_variable[columna] = 'Cualitativa Nominal'
        elif df[columna].dtype in ['int64', 'float64']:
            if unique_values < 10:
                tipo_variable[columna] = 'Cuantitativa Discreta'
            else:
                tipo_variable[columna] = 'Cuantitativa Continua'
        else:
            tipo_variable[columna] = 'Otro tipo'

    return tipo_variable

# Clasificar las variables del DataFrame
clasificacion = clasificar_tipo_variable(df)

# Mostrar resultados
print("Clasificación de las variables:")
for var, tipo in clasificacion.items():
    print(f"{var}: {tipo}")

```

```python
import pandas as pd

# Cargar la base de datos Titanic desde un repositorio en línea
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df1 = pd.read_csv(url)

# Manejo de valores faltantes en 'Age'
if df1['Age'].isnull().sum() > 0:
    df1['Age'].fillna(df1['Age'].median(), inplace=True)

# Manejo de valores faltantes en 'Embarked'
if df1['Embarked'].isnull().sum() > 0:
    if df1['Embarked'].isnull().sum() > 2:
        df1['Embarked'].fillna('S', inplace=True)
    elif df1['Embarked'].isnull().sum() <= 2:
        df1.dropna(subset=['Embarked'], inplace=True)

# Verificar las columnas existentes en el DataFrame
print("Columnas actuales en el DataFrame:", df1.columns)

# Convertir variables categóricas a dummies solo si las columnas existen
categorical_columns = ['Sex', 'Embarked']
missing_columns = [col for col in categorical_columns if col not in df1.columns]

if missing_columns:
    print(f"Las siguientes columnas faltan en el DataFrame y no pueden convertirse en dummies: {missing_columns}")
else:
    df1 = pd.get_dummies(df1, columns=categorical_columns, drop_first=True)
    print("Variables categóricas convertidas a dummies exitosamente")

# Mostrar los primeros datos después de la transformación
df1.head()
```
